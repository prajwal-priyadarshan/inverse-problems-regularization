```python
import json
from pathlib import Path
from src.llm_reasoning.diagnostic_formatter import format_diagnostics
from src.llm_reasoning.llm_interface import generate_decision, has_api_key

root = Path(__file__).resolve().parents[1]
metrics_dir = root / 'results' / 'metrics'
summaries = []
for level in ['low', 'medium', 'high']:
    summaries.append(json.loads((metrics_dir / f'summary_{level}.json').read_text()))
text = format_diagnostics(summaries)
if has_api_key():
    decision = generate_decision(text)
    print(decision)
else:
    print('OPENAI_API_KEY not set; skipping LLM call.')
```
